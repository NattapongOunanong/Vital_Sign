```python
import requests, time
from json import dumps
```

```python
url = 'http://127.0.0.1:5000/bloodPressure'
requests.get(url=url).content
requests.post(url=url, data={"command":"start"}).content
time.sleep(10)
requests.post(url=url, data={"command":"stop"}).content
```

```python
url = 'http://127.0.0.1:5000/thermalGun'
parameter={'status':True, 'temperature':1.0, 'timestamp':1}
requests.get(url=url, params=parameter).content
requests.post(url=url, data=dumps(parameter)).content
```
